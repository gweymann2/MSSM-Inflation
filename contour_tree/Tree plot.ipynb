```python
def alpha(file, dtype=float):
    if dtype == float:
        return pd_to_array(file, 'A6', dtype=float)**2/(40*pd_to_array(file, 'mphi', dtype=float)**2)
    else:
        return pd_to_array(file, 'A6', dtype='mp')**mp.mpf('2')/(mp.mpf(40)*pd_to_array(file, 'mphi', dtype='mp')**mp.mpf('2'))

def alphaB(file, dtype=float):
    list_alpha = []
    for alpha_ in alpha(file, dtype):
        list_alpha.append(0.5*mp.sqrt(1-alpha_))
    return np.array(list_alpha)
    
plt.figure(1)
plt.plot(pd_to_array(file100, 'phi0B', dtype = 'mp'),alphaB(file100, dtype = 'mp')-alphaB(file101, dtype = 'mp'),label=r'$ln(R_{rad}) = -10$', color='blue')
plt.plot(pd_to_array(file110, 'phi0B', dtype = 'mp'),alphaB(file110, dtype = 'mp')-alphaB(file111, dtype = 'mp'),label=r'$ln(R_{rad}) = 0$', color='red')
plt.plot(pd_to_array(file120, 'phi0B', dtype = 'mp'),alphaB(file120, dtype = 'mp')-alphaB(file121, dtype = 'mp'),label=r'$ln(R_{rad}) = 10$', color='green')
plt.plot(pd_to_array(file100, 'phi0B', dtype = 'mp'),alphaB(file102, dtype = 'mp')-alphaB(file101, dtype = 'mp'), color='blue', linestyle=':')
plt.plot(pd_to_array(file110, 'phi0B', dtype = 'mp'),alphaB(file112, dtype = 'mp')-alphaB(file111, dtype = 'mp'), color='red', linestyle=':')
plt.plot(pd_to_array(file120, 'phi0B', dtype = 'mp'),alphaB(file122, dtype = 'mp')-alphaB(file121, dtype = 'mp'), color='green', linestyle=':')
plt.xlabel(r'$\phi_0$ (GeV)')
plt.ylabel(r'$\alpha^{ns}-\alpha^{0.9653}$')
plt.legend(fontsize=25)
define_plot_resolution()
```

```python
lnMpinGev = mp.mpf('42.334')
Mp = mp.exp(lnMpinGev)

def Vtree(phi, mphi_GUT, A6_GUT, lambda6_GUT):
    return mp.fadd(mp.fsub(mp.fmul(mp.fmul(0.5,mp.power(mphi_GUT,2)),mp.power(phi,2)),mp.fmul(mp.fdiv(mp.fmul(lambda6_GUT,A6_GUT),(mp.fmul(6,mp.power(Mp,3)))),mp.power(phi,6))),mp.fmul(mp.fdiv(mp.power(lambda6_GUT,2),mp.power(Mp,6)),mp.power(phi,10)))

def DV(V, phi):
     return mp.fdiv((mp.fsub(V(phi+1),V(phi-1))),2)
    
def eps1_(V, phi):
    return mp.fdiv(mp.power((mp.fmul((mp.mpf(1)*Mp),mp.fdiv(DV(V, phi),V(phi)))),2),2)

def phi_end(file, tree_or_rge):
    phi_end_array = []
    n = len(pd_to_array(file, 'mphi', dtype='mp'))
    if tree_or_rge == 'tree':
        for i in range(n):
            V = lambda phi : Vtree(phi, pd_to_array(file, 'mphi', dtype='mp')[i], pd_to_array(file, 'A6', dtype='mp')[i], pd_to_array(file, 'lambda6', dtype='mp')[i])
            phi_end_array.append(mp.findroot(lambda phi: eps1_(V, phi)-1, 0.98*pd_to_array(file, 'phi0B', dtype='mp')[i], tol = 1e-30))
        return np.array(phi_end_array)
    else:
        for i in range(n):
            V = lambda phi : Vrge(phi, 1, pd_to_array(file, 'mphi', dtype='mp')[i], pd_to_array(file, 'A6', dtype='mp')[i], pd_to_array(file, 'lambda6', dtype='mp')[i])
            phi_end_array.append(mp.findroot(lambda phi: eps1_(V, phi)-1, 0.98*pd_to_array(file, 'phi0B', dtype='mp')[i], tol = 1e-30))
        return np.array(phi_end_array)

plt.figure()
plt.plot(pd_to_array(file101, 'phi0B'), pd_to_array(file101, 'phi0B')-phi_end(file101, 'tree'),'blue')
plt.fill_between([], [], [], color = 'blue', alpha=0.2, label=r'$ln(R_{rad}) = -10$')
plt.plot(pd_to_array(file111, 'phi0B'), pd_to_array(file111, 'phi0B')-phi_end(file111, 'tree'),'red')
plt.fill_between([], [], [], color = 'red', alpha=0.2, label=r'$ln(R_{rad}) = 0$')
plt.plot(pd_to_array(file121, 'phi0B'), pd_to_array(file121, 'phi0B')-phi_end(file121, 'tree'),'green')
plt.fill_between([], [], [], color = 'green', alpha=0.2, label=r'$ln(R_{rad}) = 10$')
plt.xlabel(r'$\phi_0$ (GeV)')
plt.ylabel(r'$\phi_0-\phi_{end}$')
plt.legend(fontsize=25)
define_plot_resolution()

Pstar = 2.10310517e-9
def Nstar(file, tree_or_rge, lnRrad):
    res = []
    n = len(pd_to_array(file, 'mphi', dtype='mp'))
    for i in range(n):
        V = lambda phi : Vtree(phi, pd_to_array(file, 'mphi', dtype='mp')[i], pd_to_array(file, 'A6', dtype='mp')[i], pd_to_array(file, 'lambda6', dtype='mp')[i])
        phi_end = mp.findroot(lambda phi: eps1_(V, phi)-1, 0.98*pd_to_array(file, 'phi0B', dtype='mp')[i], tol = 1e-30)
        kstar, lnMpcToKappa, HubbleSquareRootOf3OmegaRad, RelatDofRatio = 0.05, 130.282, 7.5437e-63, 1
        N0 = log(kstar) - lnMpcToKappa - 0.5*log(HubbleSquareRootOf3OmegaRad) -0.25*log(RelatDofRatio)
        Nstar = lnRrad - N0 - 0.25*mp.log(9/eps1_(V, phi)/(3-eps1_(V, phi_end))*V(phi_end)/V(phi))+0.25*mp.log(8*mp.pi**2*Pstar)
        res.append(float(Nstar))
    return np.array(res)

plt.figure()
plt.plot(pd_to_array(file101, 'phi0B', -10), Nstar(file101,'tree', -10),'blue')
plt.plot(pd_to_array(file111, 'phi0B', 0), Nstar(file111,'tree', 0),'red')
plt.plot(pd_to_array(file121, 'phi0B', 10), Nstar(file121,'tree', 10),'green')
plt.fill_between([],[],[], color = 'blue', alpha=0.2, label=r'$ln(R_{rad}) = -10$')
plt.fill_between([],[],[], color='red', alpha=0.2, label=r'$ln(R_{rad}) = 0$')
plt.fill_between([],[],[], color='green', alpha=0.2, label=r'$ln(R_{rad}) = 10$')
# plt.fill_between(pd_to_array(file120, 'phi0B', 10), Nstar(file120,'tree', 10), Nstar(file122,'tree', 10), color='green', alpha=0.2, label=r'$ln(R_{rad}) = 10$')
plt.xlabel(r'$\phi_0$ (GeV)')
plt.ylabel(r'$N^*$')
plt.legend(fontsize=25)
define_plot_resolution()

def rhoend(file, tree_or_rge):
    res = []
    n = len(pd_to_array(file, 'mphi', dtype='mp'))
    if tree_or_rge == 'tree':
        for i in range(n):
            V = lambda phi : Vtree(phi, pd_to_array(file, 'mphi', dtype='mp')[i], pd_to_array(file, 'A6', dtype='mp')[i], pd_to_array(file, 'lambda6', dtype='mp')[i])
            phi_end = mp.findroot(lambda phi: eps1_(V, phi)-1, 0.98*pd_to_array(file, 'phi0B', dtype='mp')[i], tol = 1e-30)
            res.append(float(V(phi_end)))
        return np.array(res) 
    else:
        for i in range(n):
            V = lambda phi : Vrge(phi, 1, pd_to_array(file, 'mphi', dtype='mp')[i], pd_to_array(file, 'A6', dtype='mp')[i], pd_to_array(file, 'lambda6', dtype='mp')[i])
            phi_end =mp.findroot(lambda phi: eps1_(V, phi)-1, 0.98*pd_to_array(file, 'phi0B', dtype='mp')[i], tol = 1e-30)
            res.append(float(V(phi_end)))
        return np.array(res) 

plt.figure()
plt.plot(pd_to_array(file101, 'phi0B'), rhoend(file101,'tree'),'blue')
plt.plot(pd_to_array(file111, 'phi0B'), rhoend(file111,'tree'),'red')
plt.plot(pd_to_array(file121, 'phi0B'), rhoend(file121,'tree'),'green')
plt.fill_between(pd_to_array(file101, 'phi0B'), rhoend(file100,'tree'), rhoend(file102,'tree'), color = 'blue', alpha=0.2, label=r'$ln(R_{rad}) = -10$')
plt.fill_between(pd_to_array(file111, 'phi0B'), rhoend(file110,'tree'), rhoend(file112,'tree'), color='red', alpha=0.2, label=r'$ln(R_{rad}) = 0$')
plt.fill_between(pd_to_array(file121, 'phi0B'), rhoend(file120,'tree'), rhoend(file122,'tree'), color='green', alpha=0.2, label=r'$ln(R_{rad}) = 10$')
plt.xlabel(r'$\phi_0$ (GeV)')
plt.ylabel(r'$\rho_{end}$')
plt.legend(fontsize=25)
plt.loglog()
define_plot_resolution()
```
